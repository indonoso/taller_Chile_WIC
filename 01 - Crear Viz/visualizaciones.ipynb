```python
import pandas as pd
import altair as alt
```

```python
df = pd.read_csv("dataset_ejemplos.csv")
```

```python
df
```

```python
color_genero = alt.Color('genero', title='Género', scale=alt.Scale(domain=['M', 'H'], range=['#8624F5', '#1FC3AA']))
```

# ¿Hay una relación entre cuánto hablan los convencionales con la cantidad de veces que aparecen en los medios?

```python
df['n_noticias'] = df['n_noticias_HH'] + df['n_noticias_HM'] + df['n_noticias_MM']
```

```python
chart1 = alt.Chart(df).mark_point(filled=True, size=100).encode(x=alt.X('segundos_discursos', title='Tiempo de discursos'), 
                                  y=alt.Y('n_noticias', title="Cantidad de noticias en que aparece"), 
                                  color=color_genero, 
                                  tooltip='nombre')
chart1
```

# ¿Cuál es la efectividad del voto de hombres versus mujeres?

```python
chart2 = alt.Chart(df).mark_boxplot().encode(y=alt.Y('genero', title=None), x=alt.X('efectividad_voto', title='Efectividad del voto'), color=color_genero)
chart2
```

# ¿Las mujeres piden más o menos la palabra?

```python
chart3 = alt.Chart(df).mark_point(filled=True, size=100).encode(x=alt.X('segundos_discursos', title='Tiempo en segundos'), 
                                           y=alt.Y('n_discursos', title='Cantidad de discursos'), 
                                           color=color_genero, tooltip='nombre')
chart3
```

# ¿Cómo es la aparición en los medios de comunicación?

```python
df_noticias = df.melt(value_vars=['n_noticias_HH', 'n_noticias_HM', 'n_noticias_MM'], 
                      id_vars=['nombre', 'genero', 'n_discursos', 'segundos_discursos', 'efectividad_voto', 'n_noticias'], 
                     var_name='tipo_noticia', 
                     value_name='cantidad_apariciones')
```

```python
df_noticias
```

```python
df_noticias['tipo_noticia'] = df_noticias['tipo_noticia'].str.replace('n_noticias_', '')
df_noticias['tipo_noticia'] = df_noticias['tipo_noticia'].str.replace('HH', 'Un género').str.replace('MM', 'Un género').str.replace('HM', 'Ambos Géneros')
```

```python
chart4 = alt.Chart(df_noticias).mark_bar().encode(x=alt.X('genero', title=None), 
                                                  y=alt.Y('sum(cantidad_apariciones)', title='Cantidad de apariciones'), 
                                                  facet=alt.Facet('tipo_noticia', title='Géneros por noticia'), color=color_genero)
chart4
```
